```python
# Let's import the needed packages
import pandas as pd
import requests
import json
```

```python
# Let's download the json data from the forbes website
url = 'https://www.forbes.com/billionaires/page-data/index/page-data.json'
r = requests.get(url)
```

```python
# Let's save the data as json file
with open('forbes_billionaires.json','w',encoding='utf') as f:
        json.dump(r.json(), f, ensure_ascii=False, indent=4)
```

```python
def get_item(item, name):
    try: 
        item_name = item[name]
    except:
        item_name = ''
    return item_name
```

```python
# Let's get the data from the json with countries
json_data = r.json()
billionaires = []
for i,item in enumerate(json_data['result']['pageContext']['tableData']):
    rank = get_item(item, 'rank')
    name = get_item(item, 'personName')
    age = get_item(item, 'age')
    country = get_item(item, 'country')
    net_worth = get_item(item, 'netWorth')
    category = get_item(item, 'category')
    source = get_item(item, 'source')
    billionaires.append([rank,name,age,country,net_worth,category,source])
```

```python
billionaires_data = pd.DataFrame(
    billionaires,
    columns=['rank','person_name','age','country','net_worth(billion $)','category','source']
    )
```
