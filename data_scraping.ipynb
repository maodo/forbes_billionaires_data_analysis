```python
# Let's import the needed packages
import pandas as pd
import requests
import json
```

```python
# Let's download the json data from the forbes website
url = 'https://www.forbes.com/billionaires/page-data/index/page-data.json'
r = requests.get(url)
```

```python
# Let's save the data as json file
with open('forbes_billionaires.json','w',encoding='utf') as f:
        json.dump(r.json(), f, ensure_ascii=False, indent=4)
```

```python
# # Let's get the data from the json without
file_path = './forbes_billionaires.json'
with open(file_path, 'r') as f:
    #load json data from file
    data = json.load(f)
extracted_data = data['result']['pageContext']['tableData']
```

```python
data['result']['pageContext']['tableData'][0]
```

```python
def get_item(item, name):
    try: 
        item_name = item[name]
    except:
        item_name = ''
    return item_name
```

```python
# Let's get the data from the extracted data
billionaires = []
for i,item in enumerate(extracted_data):
    rank = get_item(item, 'rank')
    name = get_item(item, 'personName')
    age = get_item(item, 'age')
    country = get_item(item, 'country')
    net_worth = get_item(item, 'netWorth')
    category = get_item(item, 'category')
    source = get_item(item, 'source')
    image = get_item(item, 'squareImage')
    birthDate = get_item(item, 'birthDate')
    gender = get_item(item, 'gender')
    organization = get_item(item, 'organization')
    title = get_item(item, 'title')
    philanthropyScore = get_item(item, 'philanthropyScore')
    selfMade = get_item(item, 'selfMade')
    billionaires.append([rank,name,age,country,net_worth,category,source,image,birthDate,gender,organization,title,philanthropyScore,selfMade])
```

```python
billionaires_data = pd.DataFrame(
    billionaires,
    columns=['rank','person_name','age','country','net_worth(billion $)','category','source','image_url','birth_date','gender','organization','title','philanthropy_score','self_made']
    )
```
